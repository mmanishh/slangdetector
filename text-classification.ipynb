```python
df_toxic = pd.read_csv('new.csv')

df_toxic=df_toxic.drop(df_toxic.columns[[0,1]],1)

```

### Let's do real training

```python
X_train, X_test, Y_train, Y_test = train_test_split(df_toxic['comment_text'], 
                                                    df_toxic['is_toxic'], 
                                                    random_state=0)
X_train.shape, X_test.shape
```

```python
# extract features
vectorizer = CountVectorizer(ngram_range=(1, 2)).fit(X_train)
X_train_vectorized = vectorizer.transform(X_train)
# X_train_vectorized.toarray().shape
```

```python
# create Naive Bayes model
model = MultinomialNB(alpha=0.1)
model.fit(X_train_vectorized, Y_train)
```

```python
# Calculate True Positive Rate vs False Positive Rate
predictions = model.predict(vectorizer.transform(X_test))
score = roc_auc_score(Y_test, predictions)
score
```
